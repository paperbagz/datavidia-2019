```python
#changing visited_city and log trans into usable
temp = []
for value in flight["log_transaction"]:
    a = value.split(',')
    to_float = lambda x : [float(p) for p in x]
    a = to_float(a)
    temp.append(a)
flight["log_transaction"] = temp

temp = []
for value in test["log_transaction"]:
    a = value.split(',')
    to_float = lambda x : [float(p) for p in x]
    a = to_float(a)
    temp.append(a)
test["log_transaction"] = temp

temp = []
for value in flight["visited_city"]:
    a = value.split(',')
    temp.append(a)
flight["visited_city"] = temp

temp = []
for value in test["visited_city"]:
    a = value.split(',')
    temp.append(a)
test["visited_city"] = temp
```

```python
#jumlah kota dikunjungi
temp = []
start = []
end = []
for i in flight["visited_city"]:
    temp.append(len(i))
    start.append(i[0])
    end.append(i[-1])

num_trans = []
mean_trans = []
for i in flight["log_transaction"]:
    num_trans.append(len(i))
    mean_trans.append(sum(i)/len(i))

is_alone = []

for i in flight["no_of_seats"]:
    if i > 1:
        is_alone.append(0)
    else:
        is_alone.append(1)

flight["is_alone"] = is_alone    
flight["num_visited"] = temp
flight["start"]= start
flight["end"] = end
flight["num_trans"] = num_trans
flight["mean_trans"] = mean_trans
```
